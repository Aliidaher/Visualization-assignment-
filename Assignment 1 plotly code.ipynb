```python
import pandas as pd
import plotly.express as px

# Load dataset
df = pd.read_csv("https://linked.aub.edu.lb/pkgcube/data/551015b5649368dd2612f795c2a9c2d8_20240902_115953.csv")

print(df.columns)

```

```python
import plotly.express as px

fig = px.bar(df, x="Town", y="Total number of hotels", title="Total Number of Hotels by Town")
fig.show()
```

```python
fig = px.scatter(
    df,
    x="Total number of restaurants",
    y="Tourism Index",
    hover_name="Town",
    title="Tourism Index vs. Number of Restaurants"
)
fig.show()
```

```python
fig = px.histogram(
    df,
    x="Total number of guest houses",
    nbins=10,
    title="Distribution of Guest Houses across Towns"
)
fig.show()
```

```python
fig = px.bar(
    df,
    x="Town",
    y=["Total number of restaurants", "Total number of cafes"],
    barmode='group',
    title="Restaurants vs. Cafes by Town"
)
fig.show()
```

```python
fig = px.bar(
    df,
    x="Town",
    y=["Total number of hotels", "Total number of restaurants", "Total number of cafes", "Total number of guest houses"],
    barmode="stack",
    title="Stacked Bar: Total Hospitality Infrastructure per Town"
)
fig.show()
```
